## Part 1: Loading the Dataset; Preliminary Analysis

```python
diamonds=pd.read_csv('diamonds.txt',sep='\t')
diamonds.head(10)
```

Now, we will determine the size of the dataset.

```python
diamonds.shape
```

We will now analyze the distribution of the columns in the diamonds DataFrame.

```python
diamonds.describe()
```

## Part 2: Filtering and Sorting
We will start by viewing information about the 5 most expensive diamonds in the dataset.

```python
diamonds.loc[:,['price','carat','cut','color','clarity']].sort_values(by='price',ascending=False).head(5)
```

Now we will view information about the 5 least expensive diamonds in the dataset.

## Part 3: Working with Categorical Variables
In this section, we will be creating lists to identify the order of each of the 3 categorical variables: cut, color, and clarity.

```python
clarity_levels=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
cut_levels=['Fair','Good','Very Good','Premium','Ideal']
color_levels=['J','I','H','G','F','E','D']
```

We will now use the created categorical lists to inform Pandas of the correct order for the levels of the categorical variables.

```python
diamonds.clarity=pd.Categorical(diamonds.clarity,clarity_levels)
diamonds.cut=pd.Categorical(diamonds.cut,cut_levels)
diamonds.color=pd.Categorical(diamonds.color,color_levels)
```

Now, we will create lists of named colors to be used as visualizations later on in this project.

```python
clarity_pal=['black','indianred','chocolate','gold','darkkhaki','yellowgreen','antiquewhite','lavender']
color_pal=['slategrey','lightsteelblue','royalblue','blue','navy','darkorchid','lavenderblush']
cut_pal=['mediumslateblue','blueviolet','mediumorchid','violet','magenta','deeppink','crimson']
```

## Part 4: Displaying Counts for Categorical Variables
We will start by counting the number of diamonds at each level of cut.

```python
diamonds.cut.value_counts().sort_index()
```

We will now count the number of diamonds for each level of color.

```python
diamonds.color.value_counts().sort_index()
```

We will now count the number of diamonds for each level of clarity.

```python
diamonds.clarity.value_counts().sort_index()
```

## Part 5: Scatterplots of Price Against Carat
In this sections, we will analyze the relationship between price and carat using scatter plots

```python
plt.figure(figsize=[8,6])
plt.scatter(x=diamonds.carat,y=diamonds.price,s=20,alpha=0.4,color='deeppink')
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')
plt.show()
```

Now we will reproduce the plot above, but color the points according to clarity.

```python
plt.figure(figsize=[8,6])
for i in range(0,len(clarity_levels)):
        current_clarity=clarity_levels[i]
        sel=diamonds.loc[:,'clarity']==current_clarity
        temp_carat= diamonds.loc[sel,'carat']
        temp_price= diamonds.loc[sel,'price']
           
        plt.scatter(x=temp_carat,y=temp_price, s=20,alpha=0.4,color=clarity_pal[i],label=current_clarity.title())
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')        
plt.legend()
plt.show()
    
```

Now we will separate the points for each of the eight clarity levels into it's own subplot.

```python
plt.figure(figsize=[12,6])
for i in range(0,len(clarity_levels)):
    plt.subplot(2,4,i+1)
    current_clarity=clarity_levels[i]
    sel=diamonds.loc[:,'clarity']==current_clarity
    temp_carat= diamonds.loc[sel,'carat']
    temp_price= diamonds.loc[sel,'price']
           
    plt.scatter(x=temp_carat,y=temp_price, s=20,alpha=0.4,color=clarity_pal[i])
    plt.xlabel('Carat Size')
    plt.ylabel('Price (in $)')
    plt.title(current_clarity.title())
    plt.xlim([0,4])
    plt.ylim([0,20000])

plt.tight_layout()
plt.show()
```

## Part 6: Applying Logarithmic Transformations
We will start by adding two new columns to the diamonds dataframe in order to store the transformed variables.

```python
diamonds.loc[:,'ln_carat']=np.log(diamonds.carat)
diamonds.loc[:,'ln_price']=np.log(diamonds.price)
```

Now we will use histograms to explore the distribution of diamond prices and the log of the diamond prices.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.hist(x=diamonds.price,edgecolor='k',color='deeppink',bins=20)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Prices')
plt.subplot(1,2,2)
plt.hist(x=diamonds.ln_price,edgecolor='k',color='deeppink',bins=20)
plt.xlabel('Natural Log of Price')
plt.ylabel('Count')
plt.title('Histogram of Log Prices')
plt.tight_layout()
plt.show()
```

We will now use histograms to explore the distribution of the carat sizes, and log of the carat sizes.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.hist(x=diamonds.carat,edgecolor='k',color='darkviolet',bins=20)
plt.xlabel('Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Carat Sizes')
plt.subplot(1,2,2)
plt.hist(x=diamonds.ln_carat,edgecolor='k',color='darkviolet',bins=20)
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Log-Carat Sizes')
plt.tight_layout()
plt.show()
```

## Part 7: Scatterplot of Transformed Variables
Now we will create a scatterplot of the natural logarithm of price against the natural logarithm of carat with points colored according to their clarity. 

```python
plt.figure(figsize=[8,6])
for i in range(0,len(clarity_levels)):
        current_clarity=clarity_levels[i]
        sel=diamonds.loc[:,'clarity']==current_clarity
        temp_carat= diamonds.loc[sel,'ln_carat']
        temp_price= diamonds.loc[sel,'ln_price']
           
        plt.scatter(x=temp_carat,y=temp_price, s=20,alpha=0.4,color=clarity_pal[i],label=current_clarity.title())
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Natural Log of Price')
plt.title('Relationship between Log-Price and Log-Carat Size')        
plt.legend()
plt.show()
```

## Part 8: Grouping by Cut
Now we will create a grouped DataFrame displaying the mean price and carat size for each cut level.

```python
gb_cut=diamonds.loc[:,['cut','price','carat']].groupby('cut').mean()
gb_cut
```

Now we will use bar charts to graphically display the information from the dataframe above.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.bar(x=gb_cut.index,height=gb_cut.price,edgecolor='k',color=cut_pal)
plt.xlabel('Cut')
plt.ylabel('Mean Price')
plt.title('Mean Price by Cut')
plt.subplot(1,2,2)
plt.bar(x=gb_cut.index,height=gb_cut.carat,edgecolor='k',color=cut_pal)
plt.xlabel('Cut')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Cut')
plt.show()
```

## Part 9: Grouping by Color
Now we will create a grouped dataframe displaying the mean price and carat size for each color level.

```python
gb_color=diamonds.loc[:,['color','price','carat']].groupby('color').mean()
gb_color
```

We will now use bar charts to graphically display the information from the dataframe above.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.bar(x=gb_color.index,height=gb_color.price,edgecolor='k',color=color_pal)
plt.xlabel('Color')
plt.ylabel('Mean Price')
plt.title('Mean Price by Color')
plt.subplot(1,2,2)
plt.bar(x=gb_color.index,height=gb_color.carat,edgecolor='k',color=color_pal)
plt.xlabel('Color')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Color')
plt.show()
```

Finally, we will create a grouped dataframe displaying the mean price and carat size for each clarity level.
